```python
import xml.etree.ElementTree as ET
import numpy as np
tree1 = ET.parse('Gleiser_comics_weighted_network-1.xml')
tree2 = ET.parse('Gleiser_comics_weighted_network-2.xml')
tree3 = ET.parse('Gleiser_comics_weighted_network-3.xml')
tree4 = ET.parse('Gleiser_comics_weighted_network-4.xml')
root1 = tree1.getroot()
root2 = tree2.getroot()
root3 = tree3.getroot()
root4 = tree4.getroot()
```

```python
f = open("marvel.graphml", "w")
f.write("<?xml version='1.0' encoding='utf-8'?>")
f.write("\n")
f.write("<graphml xmlns=\"http://graphml.graphdrawing.org/xmlns\" xmlns:xsi=\"http://www.w3.org/2001/XMLSchema-instance\" xsi:schemaLocation=\"http://graphml.graphdrawing.org/xmlns http://graphml.graphdrawing.org/xmlns/1.0/graphml.xsd\">")
f.write("\n")
f.write("<key attr.name=\"name\" attr.type=\"string\" for=\"node\" id=\"1\"/>")
f.write("\n")
f.write("<graph edgedefault=\"undirected\">")
f.write("\n")
```

```python
id_list = []

for node1 in root1.iter('node'):
    label = node1.get('id')
    name = node1.get('title')
    f.write("<node id=\"")
    f.write(label)
    f.write("\">")
    f.write("\n")
    f.write("  <data key=\"1\">")
    f.write(name)
    f.write("</data>")
    f.write("\n")
    f.write("</node>")
    f.write("\n")
    id_list.append(label)
```

```python
for node2 in root2.iter('node'):
    if not str(node2.get('id')) in id_list:
        label = node2.get('id')
        name = node2.get('title')
        f.write("<node id=\"")
        f.write(label)
        f.write("\">")
        f.write("\n")
        f.write("  <data key=\"1\">")
        f.write(name)
        f.write("</data>")
        f.write("\n")
        f.write("</node>")
        f.write("\n")
        id_list.append(label)
```

```python
for node3 in root3.iter('node'):
    if not str(node3.get('id')) in id_list:
        label = node3.get('id')
        name = node3.get('title')
        f.write("<node id=\"")
        f.write(label)
        f.write("\">")
        f.write("\n")
        f.write("  <data key=\"1\">")
        f.write(name)
        f.write("</data>")
        f.write("\n")
        f.write("</node>")
        f.write("\n")
        id_list.append(label)
```

```python
for node4 in root4.iter('node'):
    if not str(node4.get('id')) in id_list:
        label = node4.get('id')
        name = node4.get('title')
        f.write("<node id=\"")
        f.write(label)
        f.write("\">")
        f.write("\n")
        f.write("  <data key=\"1\">")
        f.write(name)
        f.write("</data>")
        f.write("\n")
        f.write("</node>")
        f.write("\n")
        id_list.append(label)
        
```

```python
array1 = np.array = []
array2 = np.array = []
array3 = np.array = []
array4 = np.array = []

linklist1 = []
linklist2 = []
linklist3 = []
linklist4 = []

edgelist1 = []
edgelist2 = []
edgelist3 = []
edgelist4 = []

for link in root1.iter('link'):
    array1.append(link)
    
for link in root2.iter('link'):
    array2.append(link)
    
for link in root3.iter('link'):
    array3.append(link)
    
for link in root4.iter('link'):
    array4.append(link)
    
linklist1 = array1
linklist2 = array2
linklist3 = array3
linklist4 = array4

linklist1 = linklist1[::2]
linklist2 = linklist2[::2]
linklist3 = linklist3[::2]
linklist4 = linklist4[::2]

i = 0
for link in linklist1:
    source = linklist1[i].get('source')
    target = linklist1[i].get('target')
    pair = [source, target]
    edgelist1.append(pair)
    i += 1
    
i = 0
for link in linklist2:
    source = linklist2[i].get('source')
    target = linklist2[i].get('target')
    pair = [source, target]
    edgelist2.append(pair)
    i += 1
    
i = 0
for link in linklist3:
    source = linklist3[i].get('source')
    target = linklist3[i].get('target')
    pair = [source, target]
    edgelist3.append(pair)
    i += 1
    
i = 0
for link in linklist4:
    source = linklist4[i].get('source')
    target = linklist4[i].get('target')
    pair = [source, target]
    edgelist4.append(pair)
    i += 1
        
reduced_edgelist = edgelist1


j=0
for pair in edgelist2:
    if pair in reduced_edgelist or [pair[1], pair[0]] in reduced_edgelist:
        j += 1
    else:
        reduced_edgelist.append(pair)
        j += 1
        
j=0
for pair in edgelist3:
    if pair in reduced_edgelist or [pair[1], pair[0]] in reduced_edgelist:
        j += 1
    else:
        reduced_edgelist.append(pair)
        j += 1
        
j=0
for pair in edgelist4:
    if pair in reduced_edgelist or [pair[1], pair[0]] in reduced_edgelist:
        j += 1
    else:
        reduced_edgelist.append(pair)
        j += 1

for pair in reduced_edgelist:
    f.write("<edge source=\"")
    f.write(pair[0])
    f.write("\" ")
    f.write("target=\"")
    f.write(pair[1])
    f.write("\"/>")
    f.write("\n")
    i += 1
        
f.write("</graph></graphml>")
f.close()
```
